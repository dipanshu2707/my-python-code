## Inheritance

- Inheritance is a way of creating a new class from an existing class.

```python
class Employee:                    # Base class
    def __init__(self,name,age,id):
        self.name = name
        self.age = age
        self.id = id

    def showDetails(self):
        print(f"the name of employee:{self.name} and age is {self.age} and id is {self.id}")


class Programmer(Employee):      # Derived class
    def showLanguage(self):           
        print("the language is java") 


# e = Employee("Rohan Das",22,121)
# e.showDetails()
# e1 = Programmer("himanshu", 20, 122)
# e1.showDetails()
# e1.showLanguage()
m1 = Programmer( "himanshu", 20, 122)
print(m1.showDetails())
m2 = Programmer()
print(m2.showLanguage())
```

```python
class parent:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex
    def showDetails(self):
        print(f"the name is {self.name} and age is {self.age} and sex is {self.sex}")
class child(parent):
    def __init__(self):
        print("this is child class")
    def details(self,roll):
        print("this is details of child class",roll)
c1 = child()
print(c1.details(24))
c2 = child("manilal",20,"male")
c2.details(24)
c2.showDetails()

```

### Single Inheritance

- Single inheritance occure when chil class inherite only a single parent class.

```python
class Animal:
    def sound(self):
        print("Animal make sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks") 

d = Dog()
d.sound()
d.bark()


```

### Multiple Inheritance

- Multiple inheritance occurs when the child class inherits from more than one parent classes.

```python
class Employee:
    company = "ITC"
    name = "Animal"
    def show(self):
        print(f"the name of employee is {self.name} and the company is {self.company}")

class coder:
    language = "java"
    def printLanguage(self):
        print(f"out of all the languages here is your language{self.language}")

class Programmer(Employee,coder):
    company = "ITC Infotech"
    def showLanguage(self):                
        print(f"the name of employee is {self.company} and the language is {self.language}")

b = Programmer()
b.show()

b.printLanguage()
b.showLanguage()
```
