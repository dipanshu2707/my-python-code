## Strings

- string is a data type in python.

- A string in python is a sequence of characters (letters, numbers, symbols) enclosed in qutoes.

     Ex = a = "harry"

- write a string in these three ways.

    - a = 'harry'            # single quoted string
    - b = "harry"          # double quoted string
    - c = '''harry'''      # Triple quoted string

 ### string slicing

- A string in python can be sliced for getting a part of the string.
- string slicing means extracting a part of a string using indexes.

      - string[start:end:step]
       
    

   NAME = "harry"   
             0, 1, 2, 3, 4

           -5,-4,-3,-2,-1 

S = name [ind_start: ind_end]

s[0:3]returns "har" - characters from 0 to 3
s[1:3]returns "ar" - characters from 1 to 3 

```python
name = "himanshu singh"

print(name[0:3])
print(name[-7:-10:-1])
print(name[1:7])
print(name[:4])  # is same as print(name[1:8])
print(name[1:])  # is same as print(name[1:8])
print(name[0:8]) 
print(name[0:7])
print(name[-1:-14:-1])
print(name[::-1])

```

```python
name = "himanshu","singh"
name1 = ["himanshu","singh"]


print(name[1])
print(type(name))


print(name1[1])
print(type(name1))
```

### slicing with skip value

- we can provide a skip(jump) value as a part of our slice like this. 

```python
word = "amazing"

print(word[1:6:2]) 
print(word[:7])  # word [0:7] 'amazing'
print(word[::2])
```

### String functions

```python
name = "dipanshu singh"

print(len(name))
print(name.endswith("hu"))
print(name.startswith("di"))
print(name.capitalize())
print(name.lower())
print(name.upper())
print(name.find("singh"))
print(name.replace("singh","kumar"))
  
```

#### Escape sequence characters

-sequence of characters after backslash "\"       
-escape sequence characters comprise of more than one character but represent one character when used within the string.
   
     -example \n,\t,\;,\\etc.

```python
a =  "hello word"

print("hello\nworld")   # New line
print("hello\tworld")   # tab (space)
print("hello\\world")   # backslash
print("hello\'world")   # single quote
print("hello\"world\"") # double quote

print("hello wor\bd")   # remove one character
print("hello\rword")    #carring return
print("hellowor\rld")
```
