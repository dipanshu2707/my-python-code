#### Modes of opening a file

r - open for reading

w - open for writing

a - open for appending

'+' -open for updating

'rb' will open for read in binary mode

'rt' will open for read in text mode

4. A file continue a word "Donkey" multiple times. you need to write a program which replace this word with ### by update the same file.
