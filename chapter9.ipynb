### File I/O

- A file is data stored on a storage device(like HDD).
- A python program cam talk to the file by reading content from it and writing content to it.
          
          -- RAM (Volatile memory)
            Temporary storage data lost after program ends.

           - HDD(NON-volatile memort)
            Permanent storage, data remains saved.

#### Modes of opening a file

r - open for reading

w - open for writing

a - open for appending

'+' -open for updating

'rb' will open for read in binary mode

'rt' will open for read in text mode

```python
st = "hey ram  you are  the best"

f = open("myfile.txt","a")
f.write(st)
f.close()
```
