4. write a recursive function to calculate the sum of first n natural numbers.

''' 
sum(1) = 1
sum(2) = 1 + 2
sum(3)=  1+2+3
sum(n) = 1+2+3+4....n-1+n
sum(n) = sum(n-1) + n
'''

```python
def sum(n):
    if(n==1):
        return 1
    return sum(n-1) + n
print(sum(5))
```

```python
write a python function to remove a given word from a list ad strip it at the same.
```

```python
def rem(l,word):
    for item in l:
        l.remove(word)
        return l
    
l = ["ram","roham","shumbham","an"]

print(rem(l,"an"))
```

```python
def rem(l,word):
    n = []
    for item in l:
        if not(item == word):
            n.append(item.strip(word))
    return n
        
    
l = ["ram","rohan","shumbham","an"]

print(rem(l,"an"))
```
