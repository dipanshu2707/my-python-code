### OBJECT ORIENTED PROGAMMING

```python
class employee:
    name = "ram"
    language = "python"
    salary = 1300000

harry = employee()
print(harry.name, harry.language)    
```

```python
class employee:
    
    language = "python"  # This is a class attribute
    salary = 1300000

harry = employee()
harry.name = "Harry"  # This is an instance attribute
print(harry.name,harry.salary, harry.language)   

rohan = employee()
rohan.name = "Rohan Roro"
print(rohan.name,rohan.salary, rohan.language)

```

#### Modelling a problem in oops

   - Noun - Class - employee
   - Adjective - Attributes - name,age,salary
   - verbs - methods - getsalary(),increment()

```python
class employee:
    
    language = "python"  # This is a class attribute
    salary = 1300000

harry = employee()
harry.language = "javaScript"  # This is an instance attribute
print(harry.salary, harry.language)   


# instance attributes take preference over class attributes during assignment & retrieval.
```

##### Self parameter

```python
class employee:
    
    language = "python"  # This is a class attribute
    salary = 1300000
  
    def getInfo(self):
        print(f"the language is {self.language}. the salary is {self.salary}")


harry = employee()
#harry.language = "javaScript"  # This is an instance attribute
harry.getInfo()  


# instance attributes take preference over class attributes during assignment & retrieval.
```
