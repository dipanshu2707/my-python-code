### Sets 

- Set is a collection of non-rrepetitive elements.
   - Ex = {2,4,1,67,9}

```python
s = {1,4,67,23,4,98,"ram"}
e = set()  #empty set


print(s,type(s))
print(e)
```

```python
s ={1,5,23,78,4}

print(len(s))
```

```python
s = {12,45,76,23,90}
s.add(27)


print(s)
```

```python
s = {12,6,534,76}
s.remove(534)

print(s)
```

```python
s = {1,2,3,4,5}
s.pop()

print(s)
```

```python
s = {8,7,6,5,4}
s.clear()

print(s)
```

```python
s ={2,3,5}
t = {1,3,7}
s.union(t)

print(s.union(t))
```

```python
s1 ={1,34,65,89}
s2 ={2,5,34,89,45}

print(s1.intersection(s2))

```

#### Properties Of Sets

1. sets are unordered
2. sets are unindexed. #cannot access element by index
3. there is no way to change items in sets.
4. Sets cannot contain duplicate values.

```python
s1 = {1,4,7,5}
s2 = {1,3,6,5}
c = s1-s2

print(c)
```
