### Tuples

- A tuple is a collection data type in python user to store multiple items in a single varible.

- Tuples are immutable (cannot be changed after creation) 

```python
a = (1, 3, 5, 7)
print(type(a)) 
```

```python
a = (1,34,452,34,False,"rohan")
no = a.count(34)
 
print(len(a))
print(no)
```

```python
a = (1,34,452,34,False,"rohan")
i = a.index(452)

print(i)
```

```python
tuple1 = (1, 3, 5)
tuple2 = (4, 6, 9)
concatenated = tuple1 + tuple2

print(concatenated)
```

```python
tuple =  (1,4,2,8,9)
repeated = tuple*2
print(repeated)
```

```python
tuple = (2,4,7,1)
print(4 in tuple) 
```

```python
my_tuple = (2,4,67,7)
sliced = my_tuple[1:3] 
print(sliced)
```
